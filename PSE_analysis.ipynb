```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

pse =  pd.read_csv("./population65200.csv")
pse = pse.rename(columns={"evolution$samples":"Samples"})
sns.set_style("darkgrid")
g = sns.scatterplot(data=pse,x="objective$fractionSpecies",y="objective$fractionLinks",hue="Samples")
g.set(xlabel= "Percentage of species preserved")
g.set(ylabel= "Percentage of links preserved")
plt.savefig("./pse.png",dpi=200)
plt.show()
g1 = pse["ratioLS"] = pse["objective$fractionLinks"]/pse["objective$fractionSpecies"] 
sns.scatterplot(data=pse,x="objective$fractionSpecies",y="ratioLS",hue="Samples")
```

```python
ose_over_70 = pd.read_csv("./population3900.csv")

g = sns.jointplot(data=ose_over_70,x="numberOfProtectionPoles",y="fractionProtected")
g.ax_joint.set(xlabel = "Protection poles", ylabel = "Fraction of landscape protected")
plt.savefig("./ose_large_links_pecies_ratio.png",dpi=200)
```
